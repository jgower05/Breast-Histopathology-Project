```python
import cv2
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
import import_ipynb
from model import Net
import torch
import torch.nn as nn
import torch.optim as optim
import pickle
```

```python
BUILD_DATA = False

class Data():
    DATA_DIR = "D:/New folder/Comp Sci/Datasets/BH Data/IDC_regular_ps50_idx5"
    PICKLE_DIR = "D:/New folder/Comp Sci/Datasets/BH Data/Data"
    NON_IDC = "0"
    IDC = "1"
    LABELS = {NON_IDC: 0, IDC: 1}
    IMG_SIZE = 40
    
    def create_training_data(self):
        for filename in os.listdir(self.DATA_DIR):
            print(filename)
            f = os.path.join(self.DATA_DIR, filename)
    
            training_data = []
            #(2) Inside that folder, load 0 and 1 images into data and their labels
            for label in self.LABELS:
                new_f = os.path.join(f, label)
                #Load images in that folder into the training_data array
                for file in tqdm(os.listdir(new_f)):
                    path = os.path.join(new_f, file)
                    image = cv2.imread(path, cv2.IMREAD_ANYCOLOR)
                    image = cv2.resize(image, (self.IMG_SIZE, self.IMG_SIZE))
                    training_data.append([np.array(image), np.eye(2)[self.LABELS[label]]])
            np.random.shuffle(training_data)
            
            #Save each array of images from each folder as a separate file
            saved_data = os.path.join(self.PICKLE_DIR, filename)
            with open(saved_data + ".pkl", "wb") as t:
                pickle.dump(training_data, t)
            
if BUILD_DATA == True:   
    data = Data()
    data.create_training_data()
```

```python
#Obtain the model
model = Net()

#Train/Test split parameters
VAL_PCT = 0.1

#Go through each pickle file and load the array
PICKLE_DIR = "D:/New folder/Comp Sci/Datasets/BH Data/Data"
for pickle_file in os.listdir(PICKLE_DIR):
    f = os.path.join(PICKLE_DIR, pickle_file)
    training_data = np.load(f, allow_pickle=True)
    
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    loss_function = nn.MSELoss()
    
    X = torch.Tensor([i[0] for i in training_data]).view(-1, 40, 40)
    X = X/255.0
    y = torch.Tensor([i[1] for i in training_data])
    
    val_size = int(len(X) * VAL_PCT)
    
    #Train/Test split
    train_X = X[:-val_size]
    train_y = y[:-val_size]
    
    test_X = X[-val_size:]
    test_y = y[-val_size:]
    
    BATCH_SIZE = 8
    EPOCHS = 3
    
    for epoch in range(EPOCHS):
        for i in tqdm(range(0, len(train_y))):
            batch_x = train_X[i].view(-1, 1, 40, 40)
            batch_y = train_y[i]

            model.zero_grad()
            outputs = model(batch_x)
            loss = loss_function(outputs, batch_y)
            loss.backward()
            optimizer.step()
    print(loss)
```
